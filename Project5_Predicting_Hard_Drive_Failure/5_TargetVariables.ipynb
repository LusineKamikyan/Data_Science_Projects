Will first add the target column for regression - remaining_useful_life column, which will be the number of days until the hard drive fails, so we can only do this for the failed hard drives 

```python
# get the serial number of hard drives that have failed
failed_hd_list = hard_drive_df[hard_drive_df.failure == 1].serial_number
```

```python
len(failed_hd_list)
```

```python
# get all the smart variable information for all days for the failed hard drives
failed_hard_drive_df = hard_drive_df[hard_drive_df.serial_number.isin(failed_hd_list)]
```

```python
failed_hard_drive_df = failed_hard_drive_df.reset_index()
```

```python
failed_hard_drive_df = failed_hard_drive_df.drop('index',axis = 1)
```

```python
failed_hard_drive_df
```

```python
# get the indices of the failure date
failure_index_df = pd.DataFrame(failed_hard_drive_df[failed_hard_drive_df.failure == 1].date)
failure_index_df = failure_index_df.rename(columns={'date':'failure_date'})
failure_date_index_list = failure_index_df.index
```

```python
failure_index_df
```

```python
failure_date_index_list
```

```python
# putting the failure date together with the hard drive df
# currently the failure_date column only contains dates in the failure date indices the 
# rest of the entries in the column are NaT
failed_hard_drive_df = pd.concat([failed_hard_drive_df,failure_index_df],axis =1)
```

```python
def fill_in_dates_before_failure_date(df, curr_ind, prev_ind):
    ''' this assigns the failure date to all the entries before the 
    last working day of the hard drive (failure_date), we will use this 
    to subtract date column from, to geth the days until the hard drive fails'''
    df.failure_date[prev_ind:curr_ind+1] = df.failure_date.iloc[curr_ind]
```

```python
prev_ind = 0
for failure_index in failure_date_index_list:
    # assigning the failure date for each row per serial_number
    fill_in_dates_before_failure_date(failed_hard_drive_df, failure_index, prev_ind)
    prev_ind = failure_index+1
```

```python
# create remaining useful life columns
# this will be the target column for regression
failed_hard_drive_df['remaining_useful_life'] = failed_hard_drive_df.failure_date-failed_hard_drive_df.date
failed_hard_drive_df.remaining_useful_life = failed_hard_drive_df.remaining_useful_life.dt.days
```

```python
failed_hard_drive_df
```

```python
# Let's also add a comlumn where if remaining_usefule life > 30 
# we will assign all those values to be 42

failed_hard_drive_df['RUL_30'] = failed_hard_drive_df.remaining_useful_life
#failed_hd_df[failed_hd_df['RUL_30'] >30].RUL_30.replace(31)
failed_hard_drive_df['RUL_30'] = failed_hard_drive_df['RUL_30'].mask(failed_hard_drive_df['RUL_30'] > 30, 42)
```

```python
failed_hard_drive_df.head(30)
```

```python
print(len(failed_hard_drive_df.columns))
print(len(set(failed_hard_drive_df.columns)))
```

```python
# save the failed_hard_drive_df
failed_hard_drive_df.to_csv('failed_hard_drive_df.csv', index = False)
```

Now we can add the target column to be used in binary classification

For the failed hard drive we will use the remaining_useful_lfie column to obtain whether the hard drive will fail the next 30 days or not (1 if failed and 0 if not).

For non failed hard drives, we will remove the last 30 day observation and the target column values will be all 0.

```python
# get all daily info of the non_failed hard drives
non_failed_hd_list=hard_drive_df[~hard_drive_df.serial_number.isin(failed_hd_list)].serial_number.unique()
```

```python
len(non_failed_hd_list)
```

```python
# get the non failed hard drive data for all smart varialbe
# for all days
non_failed_hard_drive_df = hard_drive_df[hard_drive_df.serial_number.isin(non_failed_hd_list)]
```

```python
non_failed_hard_drive_df = non_failed_hard_drive_df.reset_index()
```

```python
non_failed_hard_drive_df.date
```

```python
# since the still working hard drive haven't failed column
# fail_next_30_days will be all zeros

#non_failed_hard_drive_df['fail_next_30_days'] = 0
non_failed_hard_drive_df['fail_next_60_days'] = 0
```

```python
non_failed_hard_drive_df
```

```python
# we remove the last 30 days from each hard drive
# since for the still working hard drives we 
# don't have info whether the next 30 days it will fail
# for the last 30 day info

#non_failed_hard_drive_df = non_failed_hard_drive_df.groupby('serial_number').apply(lambda x: x[:-30]).drop(['serial_number'],axis = 1).reset_index()
non_failed_hard_drive_df = non_failed_hard_drive_df.groupby('serial_number').apply(lambda x: x[:-60]).drop(['serial_number'],axis = 1).reset_index()
```

```python
non_failed_hard_drive_df = non_failed_hard_drive_df.drop(['level_1', 'index'], axis = 1)
```

Add fail_next_30_days column to the failed_hard_drive_df dataframe
 

```python
#failed_hard_drive_df['fail_next_30_days'] = failed_hard_drive_df['remaining_useful_life'].apply(lambda x: 1 if x<=30 else 0) 
failed_hard_drive_df['fail_next_60_days'] = failed_hard_drive_df['remaining_useful_life'].apply(lambda x: 1 if x<=60 else 0) 
```
